```python
import vectorbt as vbt
import pandas as pd
import numpy as np
import datetime
import yfinance
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
```

```python
df = pd.read_csv('NSE_BANKNIFTY, 5 05-07-2022 until 25-07-2023-ML.csv')
# df.set_index('time')
df.set_index('time',inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
bullish_condn = (
    (df['close'] > df['Bullish Breakout Zone']) & 
    (df['RSI'] > 55) &
    (df['DI+']>25) &
    ((df['DI+'] - df['DI-']) >= 7 ) 
)

bearish_condn = (
    (df['close'] > df['Bearish Breakout Zone']) & 
    (df['RSI'] > 45) &
    (df['DI-']>25) &
    ((df['DI+'] - df['DI-']) >= 7 ) 
)
```

```python
entry_bullish = bullish_condn & (df['close'] > df['BLR']) & (df['close'] > df['AVWAP'])
entry_bearish = bearish_condn & (df['close'] < df['BLR']) & (df['close'] < df['AVWAP'])
```

```python
tp = 60
sl = -30
```

```python
exit_bullish =   ((df['close'] >= (df['close'].shift(1) + tp)) | 
                 (df['close'] <= (df['close'].shift(1) + sl)))
exit_bearish =  ((df['close'] <= (df['close'].shift(1) - tp)) | 
                 (df['close'] >= (df['close'].shift(1) - sl)))
```

```python
portfolio = vbt.Portfolio.from_signals(
    close=df['close'],
    entries=entry_bullish.shift(1) | entry_bearish.shift(1),
    exits=exit_bullish | exit_bearish
)
```

```python
portfolio.stats()
```

```python
portfolio.plot()
```

```python
entry_records_df = portfolio.entry_trades.records
```

pd.merge(df,entry_records_df,left_on='close',right_on='entry_price',how='outer').to_csv('close_and_entry_exit.csv')
